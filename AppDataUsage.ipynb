```python
# Import Libraries

import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt


app_details = pd.read_csv(r"C:\Users\n0ore\Desktop\Screentime_App_Details.csv")
app_ranking = pd.read_csv(r"C:\Users\n0ore\Desktop\Screentime_App_Ranking.csv")
overall_usage = pd.read_csv(r"C:\Users\n0ore\Desktop\Screentime_Overall_Usage.csv")

print(app_details)
```

```python
# Check if App_Details has any null values
app_details.isnull().sum()
```

```python
# Check the disrcriptive statistics of the data

print(app_details.describe())
```

```python
# Check how much each app has been used

data =app_details
apps = app_details['App'].unique()

plt.figure(figsize=(10, 6))

for app in apps:
    app_data = app_details[app_details['App'] == app]
    plt.bar(app_data['Date'], app_data['Usage'], label=app)

plt.xlabel('Date')
plt.ylabel('Usage')
plt.title('App Usage Comparison')
plt.legend()
plt.xticks(rotation=45)

plt.show()
```

```python
# Check the number of notifications from each app

date = app_details['Date']
notifications = app_details['Notifications']
app = app_details['App'].unique()

plt.figure(figsize=(10, 6))
for a in app:
    plt.bar(date[app_details['App'] == a], notifications[app_details['App'] == a], label=a)

plt.xlabel('Date')
plt.ylabel('Number of Notifications')
plt.title('Number of Notifications by App')
plt.legend()

plt.xticks(rotation=45)

plt.show()
```

```python
# How many times have the apps been opened?

date = app_details['Date']
times_opened = app_details['Times opened']
app = app_details['App'].unique()

plt.figure(figsize=(10, 6))
for a in app:
    plt.bar(date[app_details['App'] == a], times_opened[app_details['App'] == a], label=a)

plt.xlabel('Date')
plt.ylabel('Number of Times Opened')
plt.title('Number of Times Opened by App')
plt.legend()
plt.xticks(rotation=45)

plt.show()
```

```python
# Check the correlation between the app usage and number of notifications

notifications = app_details['Notifications']
usage = app_details['Usage']

plt.figure(figsize=(8, 6))
plt.scatter(notifications, usage, s=notifications, alpha=0.5)
plt.xlabel('Number of Notifications')
plt.ylabel('App Usage')
plt.title('Correlation Between Number of Notifications and App Usage')

z = np.polyfit(notifications, usage, 1)
p = np.poly1d(z)
plt.plot(notifications, p(notifications), color='r', linestyle='--')

plt.show()
```

```python
# There is a positive correlation between the number of notifications and app usage
```
